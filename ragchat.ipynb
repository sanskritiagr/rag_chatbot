```python
from langchain.document_loaders import TextLoader
from pypdf import PdfReader
from langchain import HuggingFaceHub
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceInstructEmbeddings
from langchain.vectorstores import FAISS
from langchain.chains import RetrievalQA, ConversationalRetrievalChain
from langchain.memory import ConversationBufferWindowMemory
```

```python
# Load pdf documents
documents_1 = ''

reader = PdfReader('/content/Computer science - Wikipedia.pdf')
for page in reader.pages:
  documents_1 += page.extract_text()

documents_1[:300]
```

```python
# Document Splitting
chunk_size = 200
chunk_overlap = 10

splitter = RecursiveCharacterTextSplitter(
    chunk_size=chunk_size,
    chunk_overlap=chunk_overlap
)
split_1 = splitter.split_text(documents_1)
split_1 = splitter.create_documents(split_1)
```

```python
from langchain.embeddings import HuggingFaceInstructEmbeddings
from InstructorEmbedding import INSTRUCTOR

instructor_embeddings = HuggingFaceInstructEmbeddings(
    model_name='hkunlp/instructor-xl',
    # model_kwargs={'device': 'cuda'},
    # encode_kwargs={'normalize_embeddings': True}
)
```

```python


# Implement embeddings
db = FAISS.from_documents(split_1, instructor_embeddings)

# Save db
db.save_local('vector store/cs')
```

```python
# Load pdf documents
documents_2 = ''

reader = PdfReader('/content/Machine learning - Wikipedia.pdf')
for page in reader.pages:
  documents_2 += page.extract_text()


documents_2[:300]
# Document Splitting
split_2 = splitter.split_text(documents_2)
split_2 = splitter.create_documents(split_2)
```

```python
db = FAISS.load_local(
    '/content/sample_data/vector store/cs', instructor_embeddings
)
```

```python
# Implement embeddings for second doc
db_2 = FAISS.from_documents(split_2, instructor_embeddings)

# Save db
db_2.save_local('vector store/ml')
```

```python
# Merge two DBs
db.merge_from(db_2)
db.save_local('vector store/cs_ml')
```

```python
# Load db
loaded_db = FAISS.load_local(
    '/content/sample_data/vector store/cs_ml', instructor_embeddings
)
```

```python
# Retrieve answer
question = 'what is oops?'

search = loaded_db.similarity_search(question, k=5)
search
```
